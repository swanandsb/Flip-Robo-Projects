```python
descrip_1=driver.find_elements_by_xpath("//p[@class='col-9 color-text-secondary my-1 pr-4']")
```

```python
descrip=[]
for i in descrip_1:
    descrip.append(i.text)
```

```python
lang_1=driver.find_elements_by_xpath("//span[@class='d-inline-block ml-0 mr-3']")
```

```python
lang=[]

if len(lang_1)==1:
    for i in lang_1:
        lang.append(i.text)
else:
    lang.append('-')
```

```python
hrefs_1=driver.find_elements_by_xpath("//h1[@class='h3 lh-condensed']/a")
```

```python
href=[]
for i in hrefs_1:
    a=i.get_attribute("href")
    href.append(a)
```

```python
href
```

```python
len(href)
```

```python
contri=[]
for i in href:
    driver.get(i)
    
    time.sleep(3)
    
    contri_1=driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[5]/div/h2/a/span")
    
    if len(contri_1)==0:
        contri_2=driver.find_elements_by_xpath("//div[@class='BorderGrid BorderGrid--spacious']/div[4]/div/h2/a/span")
        
        if len(contri_2)==1:
            for k in contri_2:
                contri.append(k.text)
        else:
            contri.append('-')
             
    else:
        for j in contri_1:
            contri.append(j.text)
```

```python
star=[]
for i in href:
    driver.get(i)
    
    time.sleep(3)

    star_1=driver.find_elements_by_xpath("//a[@class='social-count js-social-count']")
    
    if len(star_1)==0:
        star.append('-')
        
    else:
        for j in star_1:
            star.append(j.text)
```

```python
print(len(title),len(descrip),len(lang),len(href),len(contri),len(star))
```

```python
table=pd.DataFrame({})
table['Repository Title']=title[:]
table['Repository Description']=descrip[:]
table['Language Used']=lang[:]
table['Contributors count)']=contri[:]
table['Stars']=star[:]
table
```
